```python
import pandas as pd
```

```python
df_historical_data = pd.read_csv('Uefa_euros_historical_data.csv')
df_fixture = pd.read_csv('Uefa_euros_fixture.csv')
```

```python
df_fixture['home'] = df_fixture['home'].str.strip()
df_fixture['away'] = df_fixture['away'].str.strip()
```

```python
df_historical_data.sort_values('year', inplace=True)
df_historical_data
```

```python
df_historical_data['home'] = df_historical_data['home'].str.strip()
df_historical_data['away'] = df_historical_data['away'].str.strip()
df_historical_data['score'] = df_historical_data['score'].str.strip()
```

```python
df_historical_data['score'] = df_historical_data['score'].str.replace('[(a.e.t.)]', "", regex=True)
```

```python
df_historical_data['score'] = df_historical_data['score'].str.replace('(/gg)', "", regex=True)
```

```python
df_historical_data['score'] = df_historical_data['score'].str.replace('(/sg)', "", regex=True)
```

```python
df_historical_data
```

```python
df = df_historical_data['score'].str.split('', expand=True)
```

```python
df.columns
```

```python
columns_to_keep = [1, 3]
df = df[columns_to_keep]
df
```

```python
df_historical_data[['HomeGoals', 'AwayGoals']] = df
df_historical_data
```

```python
df_historical_data.drop('score', axis=1, inplace=True)
```

```python
df_historical_data
```

```python
df_historical_data.rename(columns={'home': 'HomeTeam', 'away': 'AwayTeam',
                                   'year': 'Year'}, inplace=True)
```

```python
df_historical_data
```

```python
df_historical_data = df_historical_data.astype({'HomeGoals':'int64', 'AwayGoals':'int64', 'Year':'int64'})
```

```python
df_historical_data.dtypes
```

```python
df_historical_data['TotalGoals'] = df_historical_data['HomeGoals'] + df_historical_data['AwayGoals']
```

```python
df_historical_data
```

```python
df_historical_data.to_csv('clean_Uefa_euros_historical_data.csv', index=False)
```

```python
df_fixture.to_csv('clean_Uefa_euros_fixture.csv', index=False)
```
